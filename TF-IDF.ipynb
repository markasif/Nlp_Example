```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("spam_ham.csv")
```

```python
data
```

```python
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.stem import WordNetLemmatizer
lamma = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
```

```python
corups = []
for i in range (0,len(data)):
    words = re.sub('[^a-zA-Z]',' ',data['text'][i])
    words = words.lower()
    words = words.split()
    words = [lamma.lemmatize(word)  for word in words if word not in (stopwords.words("english"))]
    words = ' '.join(words)
    corups.append(words)
    
```

```python
corups
```

### Create TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid = TfidfVectorizer(max_features=100)
X= tfid.fit_transform(corups).toarray()
```

```python
##displaying X
np.set_printoptions(edgeitems=30,linewidth=100000,
                    formatter=dict(float=lambda X : "%.3g" % X))
```

```python
X
```

## N-grams

```python
tfid = TfidfVectorizer(max_features=100,ngram_range=(2,2))
X= tfid.fit_transform(corups).toarray()
```
