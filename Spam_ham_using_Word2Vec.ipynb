```python
import pandas as pd
messages = pd.read_csv("spam_ham.csv")
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    # Remove non-ASCII characters, emojis, and special symbols
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)  # Keep only alphabets
    text = text.lower()  # Convert to lowercase
    text = text.split()  # Tokenize
    text = [lemmatizer.lemmatize(word) for word in text]  # Lemmatize
    return ' '.join(text) 
```

```python
import re
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
corpus = []
for i in range(len(messages)):
    cleaned_text = clean_text(messages["text"][i])
    if len(cleaned_text) > 0:  # Skip completely empty texts
        corpus.append(cleaned_text)
    else:
        corpus.append("empty") 
```

```python
len(corpus)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_tokens = sent_tokenize(sent)
    for sent in sent_tokens:
        words.append(simple_preprocess(sent))
```
